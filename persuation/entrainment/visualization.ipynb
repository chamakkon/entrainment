```python
import matplotlib.pyplot as plt
import json
import pandas as pd
import numpy as np
from typing import List
# サンプルデータ
def visualize_entrainment(metric, feature):
    f = open("entrainment_score.json", "r")
    entrainment_score = json.load(f)
    data = {}
    for session, score in entrainment_score.items():
        if not pd.isnull(score[feature]):
            data[session] = score[feature][metric]
    scores : List[float] = data.values()
    print(scores)
    plt.clf()
    plt.hist(scores, bins=20, edgecolor='black')
    # タイトルとラベルの設定
    plt.title('Entrainment Score')
    plt.xlabel(f'{feature} {metric} score')
    plt.ylabel('Frequency')

    # グラフを表示
    plt.savefig(f"entrainment_fig/{feature}_{metric}.png")
    return
```

```python
f = open("entrainment_score.json", "r")
entrainment_score = json.load(f)
features= entrainment_score["0"].keys()
metrics = entrainment_score["0"]["F0_MAX"].keys()
for feature in features:
    for metric in metrics:
        visualize_entrainment(metric, feature)
```
